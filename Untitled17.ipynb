# Analysing Prediction Confidence in Support Vector Machines Using Margin Distance

This notebook investigates how Support Vector Machines (SVMs) express prediction confidence through the distance of data points from the decision boundary.

Unlike probabilistic classifiers, SVMs do not naturally output probabilities. However, the **decision function** provides a useful confidence signal:

- Points **far from the decision boundary** are classified with **high confidence**
- Points **close to the decision boundary** are classified with **low confidence**
- Misclassifications often occur **near the boundary**

This notebook explores these ideas through visualisation and quantitative analysis.

## Objectives

The main objectives of this notebook are:

1. Generate a synthetic classification dataset
2. Train an SVM classifier
3. Visualise the decision boundary, margins, and support vectors
4. Analyse confidence using distance from the decision boundary
5. Identify uncertain and misclassified samples
6. Study how the regularisation parameter `C` changes the confidence behaviour

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## 1. Generate a Synthetic Dataset

A synthetic dataset is created using the `make_moons` function. This dataset is chosen because it is non-linearly separable, making it well suited for demonstrating the behaviour of an SVM with an RBF kernel.

```python
np.random.seed(42)

X, y = make_moons(n_samples=400, noise=0.22, random_state=42)

print("Feature shape:", X.shape)
print("Unique classes:", np.unique(y))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.title("Synthetic Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## 2. Split and Scale the Data

The dataset is divided into training and testing sets. Feature scaling is applied because SVM performance is sensitive to feature magnitude.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

## 3. Train the SVM Model

An SVM with an RBF kernel is trained on the scaled training data. The model learns a non-linear decision boundary.

The decision function value produced by the SVM will later be used as a measure of confidence.

```python
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_train_scaled, y_train)

y_pred = svm_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)

print(f"Test Accuracy: {accuracy:.4f}")
```

## 4. Define a Function to Plot the Decision Boundary

This helper function visualises:
- Decision regions
- Decision boundary
- Support vectors

It is useful for understanding how the SVM separates the two classes.

```python
def plot_decision_boundary(model, X_original, y, scaler, title="Decision Boundary"):
    x_min, x_max = X_original[:, 0].min() - 0.8, X_original[:, 0].max() + 0.8
    y_min, y_max = X_original[:, 1].min() - 0.8, X_original[:, 1].max() + 0.8

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 400),
        np.linspace(y_min, y_max, 400)
    )

    grid = np.c_[xx.ravel(), yy.ravel()]
    grid_scaled = scaler.transform(grid)

    Z = model.predict(grid_scaled)
    Z = Z.reshape(xx.shape)

    decision_values = model.decision_function(grid_scaled).reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.25)
    plt.contour(xx, yy, decision_values, levels=[-1, 0, 1], linewidths=[1, 2, 1], linestyles=['--', '-', '--'])

    plt.scatter(X_original[:, 0], X_original[:, 1], c=y, edgecolors='k')

    support_vectors_original = scaler.inverse_transform(model.support_vectors_)
    plt.scatter(
        support_vectors_original[:, 0],
        support_vectors_original[:, 1],
        s=120,
        facecolors='none',
        edgecolors='k',
        linewidths=1.5,
        label='Support Vectors'
    )

    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```

```python
plot_decision_boundary(
    svm_model,
    X,
    y,
    scaler,
    title="SVM Decision Boundary, Margins, and Support Vectors"
)
```

## 5. Measure Confidence Using the Decision Function

For an SVM, the `decision_function` gives the signed distance-like value from the separating boundary:

- Values near **0** indicate low confidence
- Large positive values indicate confident class 1 predictions
- Large negative values indicate confident class 0 predictions

The absolute value of the decision function is used here as a practical confidence score.

```python
decision_scores = svm_model.decision_function(X_test_scaled)
confidence_scores = np.abs(decision_scores)

print("First 10 decision scores:\n", decision_scores[:10])
print("\nFirst 10 confidence scores:\n", confidence_scores[:10])
```

## 6. Visualise the Distribution of Confidence Scores

This histogram shows how test samples are distributed according to their confidence levels.

```python
plt.figure(figsize=(8, 6))
plt.hist(confidence_scores, bins=25, edgecolor='k')
plt.title("Distribution of Confidence Scores")
plt.xlabel("Absolute Distance from Decision Boundary")
plt.ylabel("Number of Samples")
plt.show()
```

## 7. Identify Low-Confidence and High-Confidence Predictions

This step separates test samples into:
- Low-confidence samples
- High-confidence samples

These groups help us understand where the model is uncertain.

```python
low_threshold = np.percentile(confidence_scores, 20)
high_threshold = np.percentile(confidence_scores, 80)

low_conf_mask = confidence_scores <= low_threshold
high_conf_mask = confidence_scores >= high_threshold

print(f"Low-confidence threshold: {low_threshold:.4f}")
print(f"High-confidence threshold: {high_threshold:.4f}")

print("Number of low-confidence samples:", np.sum(low_conf_mask))
print("Number of high-confidence samples:", np.sum(high_conf_mask))
```

## 8. Plot Test Samples by Confidence Level

This figure highlights the most uncertain and most confident predictions in the test set.

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c='lightgray', edgecolors='none', label='Other Test Samples')
plt.scatter(X_test[low_conf_mask, 0], X_test[low_conf_mask, 1], marker='o', edgecolors='k', facecolors='none', s=120, label='Low Confidence')
plt.scatter(X_test[high_conf_mask, 0], X_test[high_conf_mask, 1], marker='s', edgecolors='k', facecolors='none', s=120, label='High Confidence')

plt.title("Low-Confidence and High-Confidence Test Samples")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

## 9. Analyse Misclassified Samples

Misclassified points are especially interesting because they often appear close to the decision boundary.

This section identifies which test samples were classified incorrectly.

```python
misclassified_mask = y_pred != y_test
num_misclassified = np.sum(misclassified_mask)

print("Number of misclassified samples:", num_misclassified)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, alpha=0.35, edgecolors='k', label='Test Samples')
plt.scatter(
    X_test[misclassified_mask, 0],
    X_test[misclassified_mask, 1],
    marker='x',
    s=120,
    linewidths=2,
    label='Misclassified Samples'
)

plt.title("Misclassified Test Samples")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

## 10. Compare Confidence for Correct and Incorrect Predictions

If the margin-based confidence idea is valid, incorrect predictions should generally have lower confidence scores than correct predictions.

```python
correct_confidence = confidence_scores[~misclassified_mask]
incorrect_confidence = confidence_scores[misclassified_mask]

plt.figure(figsize=(8, 6))
plt.hist(correct_confidence, bins=20, alpha=0.7, label='Correct Predictions', edgecolor='k')
plt.hist(incorrect_confidence, bins=20, alpha=0.7, label='Incorrect Predictions', edgecolor='k')

plt.title("Confidence Comparison: Correct vs Incorrect Predictions")
plt.xlabel("Absolute Distance from Decision Boundary")
plt.ylabel("Number of Samples")
plt.legend()
plt.show()
```

## 11. Create a Confidence Map Over the Feature Space

This visualisation shows how confidence changes across the entire feature space.

Regions far from the boundary have larger absolute decision values and are therefore more confident.

```python
def plot_confidence_map(model, X_original, y, scaler, title="Confidence Map"):
    x_min, x_max = X_original[:, 0].min() - 0.8, X_original[:, 0].max() + 0.8
    y_min, y_max = X_original[:, 1].min() - 0.8, X_original[:, 1].max() + 0.8

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 400),
        np.linspace(y_min, y_max, 400)
    )

    grid = np.c_[xx.ravel(), yy.ravel()]
    grid_scaled = scaler.transform(grid)

    decision_values = model.decision_function(grid_scaled).reshape(xx.shape)
    confidence_map = np.abs(decision_values)

    plt.figure(figsize=(8, 6))
    contour = plt.contourf(xx, yy, confidence_map, levels=30, alpha=0.85)
    plt.colorbar(contour, label='Confidence (Absolute Distance from Boundary)')
    plt.contour(xx, yy, decision_values, levels=[0], linewidths=2)

    plt.scatter(X_original[:, 0], X_original[:, 1], c=y, edgecolors='k')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
plot_confidence_map(
    svm_model,
    X,
    y,
    scaler,
    title="Confidence Map Based on Margin Distance"
)
```

## 12. Quantify Accuracy Within Confidence Groups

This section divides predictions into confidence bands and evaluates how accuracy changes across them.

```python
q1 = np.percentile(confidence_scores, 33)
q2 = np.percentile(confidence_scores, 66)

low_band = confidence_scores <= q1
mid_band = (confidence_scores > q1) & (confidence_scores <= q2)
high_band = confidence_scores > q2

low_acc = accuracy_score(y_test[low_band], y_pred[low_band])
mid_acc = accuracy_score(y_test[mid_band], y_pred[mid_band])
high_acc = accuracy_score(y_test[high_band], y_pred[high_band])

print(f"Low-confidence accuracy: {low_acc:.4f}")
print(f"Mid-confidence accuracy: {mid_acc:.4f}")
print(f"High-confidence accuracy: {high_acc:.4f}")
```

```python
bands = ['Low Confidence', 'Medium Confidence', 'High Confidence']
band_accuracies = [low_acc, mid_acc, high_acc]

plt.figure(figsize=(8, 6))
plt.bar(bands, band_accuracies)
plt.ylim(0, 1.05)
plt.title("Accuracy Across Confidence Bands")
plt.xlabel("Confidence Band")
plt.ylabel("Accuracy")
plt.show()
```

## 13. Study the Effect of the Regularisation Parameter C

The regularisation parameter `C` controls how strictly the SVM tries to separate the classes.

This experiment investigates how changing `C` affects:
- Accuracy
- Mean confidence
- Number of support vectors

```python
C_values = [0.1, 1, 10, 100]
results = []

for c in C_values:
    model = SVC(kernel='rbf', C=c, gamma='scale')
    model.fit(X_train_scaled, y_train)

    pred = model.predict(X_test_scaled)
    scores = model.decision_function(X_test_scaled)
    conf = np.abs(scores)

    acc = accuracy_score(y_test, pred)
    avg_conf = np.mean(conf)
    sv_count = len(model.support_vectors_)

    results.append((c, acc, avg_conf, sv_count))

for row in results:
    print(f"C={row[0]} | Accuracy={row[1]:.4f} | Avg Confidence={row[2]:.4f} | Support Vectors={row[3]}")
```

## 14. Visualise the Effect of C on Model Behaviour

```python
C_list = [r[0] for r in results]
acc_list = [r[1] for r in results]
avg_conf_list = [r[2] for r in results]
sv_list = [r[3] for r in results]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(C_list, acc_list, marker='o')
plt.xscale('log')
plt.title("Accuracy vs C")
plt.xlabel("C (log scale)")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(C_list, avg_conf_list, marker='o')
plt.xscale('log')
plt.title("Average Confidence vs C")
plt.xlabel("C (log scale)")
plt.ylabel("Average Confidence")
plt.show()
```

## 15. Detailed Classification Report

```python
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))
```

## 16. Summary of Findings

The analysis in this notebook shows the following:

1. The SVM decision function provides a useful confidence signal.
2. Samples close to the decision boundary tend to be more uncertain.
3. Misclassified samples generally have lower confidence than correctly classified samples.
4. Accuracy is typically higher in the high-confidence region than in the low-confidence region.
5. The regularisation parameter `C` changes not only accuracy but also the confidence structure and number of support vectors.

These findings demonstrate that margin distance is a valuable tool for interpreting SVM predictions.

## 17. Conclusion

This notebook investigated prediction confidence in Support Vector Machines using margin distance. The results showed that the absolute value of the decision function can be used as a practical confidence measure.

Points closer to the decision boundary were associated with greater uncertainty and a higher likelihood of misclassification, while points farther from the boundary were generally classified correctly with higher confidence.

Overall, the study highlights that SVMs can be interpreted not only through their decision boundaries but also through the confidence information embedded in their margin distances.
